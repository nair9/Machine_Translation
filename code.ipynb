```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset("Muennighoff/flores200",'all')
```

**1. Data preparation (5points)**

1) Data download and preprocessing

```python
for data in dataset['dev']:
  # Here I print out one example in the 'dev' set.
  print(data)
  for itm in data:
    print(itm, ':', data[itm])

  # Select the sentecen pairs based on your interest of languages.
  # The code for each language is available on this page:
  # https://github.com/facebookresearch/flores/blob/main/flores200/README.md
  # Here I select English and French. and print out the corresponding sentecens.
  print('sentence in english: \n', data['sentence_eng_Latn']) #English
  print('sentence in malayalam: \n', data['sentence_mal_Mlym']) #Malayalam
  break
```

Selecting english as the source language and malayalam as the target language

```python
source_language = 'sentence_eng_Latn' #English
target_language = 'sentence_mal_Mlym' #Malayalam
```

```python
data_eng = dataset['dev']['sentence_eng_Latn']
data_mal = dataset['dev']['sentence_mal_Mlym']
```

```python
df_eng = pd.DataFrame({'sentence_eng_Latn': data_eng})
df_mal = pd.DataFrame({'sentence_mal_Mlym': data_mal})

# Combine DataFrames
combined_df = pd.concat([df_eng, df_mal], axis=1)
```

```python
combined_df
```

```python
# Select 100 sentence pairs
num_sentences = 100
selected_data = combined_df.sample(n=100,random_state=1234).reset_index(drop=True)
```

```python
selected_data
```

**2. Machine Translation with Seq2Seq model (65 points, 20 points for each model, 5 points for
data statistics)**

1) OPUS-MT

```python
# Load necessary libraries
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, MarianMTModel
```

```python
src = "en"  # source language
trg = "ml"  # target language

mt_tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ml")
mt_model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-ml")

sample_text = "Where's the bus stop?"
print(sample_text)
batch = mt_tokenizer([sample_text], return_tensors="pt")

generated_ids = mt_model.generate(**batch)
mt_tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

```python
mt_results = []
for i in list(selected_data[source_language]):
  batch = mt_tokenizer(i, return_tensors="pt")
  generated_ids = mt_model.generate(**batch)
  ans = mt_tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
  mt_results.append(ans)
```

```python
mt_results
```

2) M2M-100

```python
# Load necessary libraries
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

m2m_tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")
m2m_model = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")
```

```python
sample_text = "Where's the bus stop?"
print(sample_text)

m2m_tokenizer.src_lang = "en"
encoded = m2m_tokenizer(sample_text, return_tensors="pt")
generated_tokens = m2m_model.generate(**encoded, forced_bos_token_id=m2m_tokenizer.get_lang_id("ml"))
m2m_tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
```

```python
m2m_results = []
for i in list(selected_data[source_language]):
  m2m_tokenizer.src_lang = "en"
  encoded = m2m_tokenizer(i, return_tensors="pt")
  generated_tokens = m2m_model.generate(**encoded, forced_bos_token_id=m2m_tokenizer.get_lang_id("ml"))
  ans = m2m_tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
  m2m_results.append(ans)
```

```python
m2m_results
```

3) MBART-50

```python
# Load necessary libraries
from transformers import AutoTokenizer, MBartForConditionalGeneration, MBart50TokenizerFast

mbart_tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
mbart_model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
sample_text = "Where's the bus stop?"
print(sample_text)

mbart_tokenizer.src_lang = "en_XX"
encoded = mbart_tokenizer(sample_text, return_tensors="pt")
generated_tokens = mbart_model.generate(
    **encoded,
    forced_bos_token_id=mbart_tokenizer.lang_code_to_id["ml_IN"]
)
mbart_tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
```

```python
mbart_results = []
for i in list(selected_data[source_language]):
  mbart_tokenizer.src_lang = "en_XX"
  encoded = mbart_tokenizer(i, return_tensors="pt")
  generated_tokens = mbart_model.generate(
      **encoded,
      forced_bos_token_id=mbart_tokenizer.lang_code_to_id["ml_IN"]
  )
  ans = mbart_tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
  mbart_results.append(ans)
```

```python
mbart_results
```

**3. Results analysis and evaluation (30 points, 10 points for each sub-question)**

```python
# importing necessary libraries
import evaluate
from nltk.translate.bleu_score import corpus_bleu
import torch

bleu = evaluate.load("bleu")
```

```python
references = list(selected_data[target_language])
```

```python
# Calculate BLEU scores
bleu = evaluate.load("bleu")
bleu_opus = bleu.compute(predictions=mt_results, references=references)
bleu_m2m = bleu.compute(predictions=m2m_results, references=references)
bleu_mbart = bleu.compute(predictions=mbart_results, references=references)

# Print BLEU scores
print(f"\nBLEU Score OPUS-MT: {bleu_opus['bleu']}")
print(f"\nBLEU Score M2M-100: {bleu_m2m['bleu']}")
print(f"\nBLEU Score MBART-50: {bleu_mbart['bleu']}")
```

**Based on the BLEU scores, the OPUS-MT has the highest BLEU score among the three models, followed by MBART-50, and M2M-100 has the lowest BLEU score.**

#### Model Performance Analysis:

1. **OPUS-MT:**
   - **BLEU Score:** 0.062
   - **Performance:** Relatively higher BLEU score compared to the other models.
   - **Possible Reasons:** OPUS-MT may perform better in terms of n-gram precision and overall similarity to reference translations. This could be attributed to the training data, model architecture, or other factors.

2. **MBART-50:**
   - **BLEU Score:** 0.034
   - **Performance:** Moderate BLEU score, falling between OPUS-MT and M2M-100.
   - **Possible Reasons:** MBART-50's performance could be influenced by its pre-training data, tokenization approach, or model design. It may have a balance between the precision of OPUS-MT and the recall of M2M-100.

3. **M2M-100:**
   - **BLEU Score:** 0.022
   - **Performance:** Lowest BLEU score among the three models.
   - **Possible Reasons:** M2M-100 may face challenges in capturing the nuances of the translation task, resulting in lower precision or recall. It could be influenced by the diversity and quality of training data or model complexity.
